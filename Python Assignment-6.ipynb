```python
import numpy as np
import pandas as pd
```

```python
# typing the information given in the assignment

df = pd.DataFrame({'From_To' : ['loNDon_paRIs', 'MAdrid_miLAN', 'londON_StockhOLM', 'BudaPest_PaRis', 'Brussels_LonDoN'], 'FlightNumber': [10045, np.nan, 10065, np.nan, 10085], 'Recent_Delays': [[23,47],[],[24,43,87],[13],[67,32]], 'Airline':  ['KLM(!)', '<Air France> (12)', '(British Airways. )',
'12. Air France', '"Swiss Air"'] })
```

```python
df
```

```python
# Filling the FlightNumber with the respective values mentioned in the assignment

df.FlightNumber = df.FlightNumber.fillna(value = df.FlightNumber[0:3].mean() , limit = 1 )
```

```python
df
```

```python
df.FlightNumber = df.FlightNumber.fillna(value = df.FlightNumber[2:].mean() , limit = 1 )
```

```python
# Final df after adding the two flight numbers
df
```

```python
# Splitting the From_To colum into 2 columns

df[['From','To']] = df.From_To.str.split(pat = '_' , expand=True) 
```

```python
df
```

```python
# Dropping the From_To column from the data frame
df = df.drop(columns = ['From_To'])
```

```python
# Lets check the final df with separated from to columns

df
```

```python
# Captalizing all the city names

df.From = df.From.str.capitalize()

df.To = df.To.str.capitalize()
```

```python
# df with capitalized names

df
```

```python
# Splitting the Recent_Delays column into seperate columns with Nan value in absent location

RDcolumns = pd.DataFrame(df.Recent_Delays.tolist(), )
```

```python
# Naming the columns to delay_1, delay_2, delay_3

RDcolumns = RDcolumns.rename(columns = {0 : 'delay_1', 1 : 'delay_2', 2: 'delay_3'})
```

```python
RDcolumns
```

```python
#joining this data frame to the previous one

df_final = df.join(RDcolumns)
```

```python
df_final
```

```python
# Deleting the Recent_Delays columsna s we don't need it

df_final = df_final.drop(columns = 'Recent_Delays')
```
